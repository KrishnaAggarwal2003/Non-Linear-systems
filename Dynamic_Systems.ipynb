# LORENTZ SYSTEM

```python
# Lorentz System
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
from mpl_toolkits.mplot3d import Axes3D

# Equation parameters
sigma=10.0
rho=28.0
beta=8/3
# Time parameters
time=np.linspace(0,100,10000)

# Derivatiev functions
def derivatives(all,t):
    x,y,z=all
    dxdt=sigma*(y-x)
    dydt=x*(rho-z)-y
    dzdt=x*y-beta*z
    return dxdt,dydt,dzdt

initials=[10.0,10.0,10.0]
sol=odeint(derivatives,initials,time)

x=sol[:, 0]
y=sol[:, 1]
z=sol[:, 2]

# 3-D Graph
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, lw=0.5)
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
plt.show()
```

```python
# Example of attractor as a point(at origin)
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Time parameters
time=np.linspace(0,100,10000)

# Derivatiev functions
def derivatives(all,t):
    x,y=all
    dxdt=-y-0.1*x
    dydt=x-0.4*y
    return dxdt,dydt

initials=[-5.0,5.0]
sol=odeint(derivatives,initials,time)

x=sol[:, 0]
y=sol[:, 1]

#Graph
plt.figure(figsize=(8, 8))
ax=plt.subplot()
ax.set_xlim([-8, 8])
ax.set_ylim([-8, 8])
ax.plot(x,y)
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
plt.show()
```

```python
# An example of Limit Cycle attractor
from typing import MutableSet
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Time parameters
time=np.linspace(0,100,10000)

mu=1
# Derivatiev functions
def derivatives(all,t):
    x,y=all
    dxdt=mu*(x - x**3/3 - y)
    dydt=x/mu
    return dxdt,dydt

initials=[-5.0,5.0]
sol=odeint(derivatives,initials,time)

x=sol[:, 0]
y=sol[:, 1]

#Graph
plt.figure(figsize=(8, 8))
ax=plt.subplot()
ax.set_xlim([-8, 8])
ax.set_ylim([-8, 8])
ax.plot(x,y)
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
plt.show()
```
